```python
import pandas as pd
 
df1 = pd.read_excel(file name)
```

```python
df1.drop('UNIQUE_ID', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler
names=['CREDIT_CARD','CASH_DEPOSIT','ATM_WITHDRAWAL','ATM_DEPOSIT','BILLS','DEBIT_CARD','OUTWARD_TRANSFER','INWARD_TRANSFER','CASH_WITHDRAWAL','CREDIT_CHEQUES','DEBIT_CHEQUES']
scaler=StandardScaler()
scaler.fit(df1)
array=scaler.transform(df1)
normalized_df = pd.DataFrame(array, columns=names)
normalized_df
```

```python
#df1.drop('cluster', axis=1, inplace=True)
df1
```

```python
from sklearn.cluster import OPTICS

#clustering = OPTICS(min_samples=7,xi=0.001,min_cluster_size=0.5).fit(df1)
clustering = OPTICS(min_samples=7,xi=0.001,min_cluster_size=0.5).fit(normalized_df)
clustering.labels_
```

```python
df1['cluster'] = pd.Series(clustering.labels_, index=df1.index)
```

```python
df1.to_excel(file name)
```

```python
from sklearn import metrics
metrics.silhouette_score(normalized_df, clustering.labels_)
```
